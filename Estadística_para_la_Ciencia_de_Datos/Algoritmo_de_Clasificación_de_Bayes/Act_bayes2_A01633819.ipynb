```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/sample_data/Social_Network_Ads.csv')
df.head()
```

```python
df = df.drop(['User ID'], axis=1)
df2 = df
df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,[1,2]], df.iloc[:,3], test_size=0.2,random_state=42)
x_train
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
```

```python
x_test
```

```python
y_pred=nb.predict(x_test)
print(y_pred)
```

```python
print("Exactitid del modelo: ", nb.score(x_test,y_test))
```
